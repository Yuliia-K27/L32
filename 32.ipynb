```python
# Створення датафрейму 15*5 з даними про студентів
import pandas as pd

students_data = {
    'Name': ['Ann', 'Bogdan', 'Viktor', 'Maryna', 'Dmytro', 'Olha', 'Danylo', 'Ivan', 'Nazar', 'Alyna', 'Olena', 'Mykola', 'Oleh', 'Lydmyla', 'Roman'],
    'Average score': [4.8, 4.2, 4.6, 4.1, 4.3, 4.2, 4.4, 4.1, 4.9, 4.7, 4.3, 4.5, 4.1, 4.6, 4.9],
    'Exam score': [85, 75, 80, 70, 77, 65, 69, 82, 87, 64, 76, 69, 73, 60, 85],
    'Number of attempts': [1, 2, 3, 2, 1, 2, 1, 1, 1, 2, 3, 2, 1, 2, 2]
     }

df_students = pd.DataFrame(students_data, index=range(1,16))
df_students
```

```python
# Додавання нової колонки "Доступ студента до іспиту" за допомогою логічної операції
df_students['Admission to the exam'] = df_students['Average score'] >= 4.5

df_students
```

```python
# Виведення мінімального значення середнього бала
print("Minimum value of the Average score: {}".format(df_students['Average score'].min()))

# Виведення максимального значення середнього бала
print("Maximum value of the Average score: {}".format(df_students['Average score'].max()))

# Виведення середнього значення середнього бала
print("Average value of the Average score: {}".format(df_students['Average score'].mean()))

# Виведення медіанного значення середнього бала
print("Median grade point Average: {}".format(df_students['Average score'].median()))
```

```python
# Виведення кількості студентів, які склали іспит з першого разу
num_passed_first_try = df_students[df_students['Number of attempts'] == 1]['Name'].count()
print("The number of students who passed the exam the first time:", num_passed_first_try)
```
